```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s = pd.Series([1,3,5, np.nan, 6,8])
print(s)
print(s[0])
print(s[0:3])
```

```python
s = pd.Series([1,3,5, np.nan, 6,8], index=['a', 'b', 'c', 'd', 'e', 'f'])
print(s['c'])
print(s[0])
```

```python
s = pd.Series({'a':1, 'b':3, 'c':5, 'd':np.nan, 'e':6, 'f':8})
print(s['c'])
```

```python
s1 = pd.Series({'a':1, 'b':3, 'c':5, 'd':np.nan, 'e':6, 'f':8})
s2 = pd.Series({'a':1, 'b':3})

s_new = s1 + s2

print(s_new)
```
