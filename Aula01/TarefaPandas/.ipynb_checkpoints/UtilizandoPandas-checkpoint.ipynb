# Utilizando Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# A tarefa 1 pedia o carregamento e a leitura do pokemon.csv agora utilizando o pandas
pokemon_csv = pd.read_csv('./pokemon.csv')
pokemon_weight = pokemon_csv['weight_kg']
# Retira os valores nulos
pokemon_weight = pokemon_weight.dropna()
```

```python
# A tarefa 3 pedia para calcular a média e o desvio padrão
media = np.mean(pokemon_weight)
desvio_padrao = np.std(pokemon_weight)
print('Média dos pesos: ', media)
print('Desvio Padrão dos pesos:', desvio_padrao)
```

```python
# A tarefa 2 pedia visualização com o matoplotlib do scatter plot
x = range(len(pokemon_weight))
scat = plt.scatter(x, pokemon_weight)
```

```python
'''
A tarefa 4 pedia para obter uma amostra, calcular a média da amostra repetindo o processo de obter a
amostra 1000 vezes, gerando um histograma das médias da amostra
'''

weight_list = []
lista_randomica = {}
lista_media = []

for line in pokemon_weight:
    weight_list.append(line)

# Separação da amostra randomica para a geração de médias para criação do gráfico
for i in range(1000):
    lista_randomica = random.sample(weight_list, 20)
    media = np.mean(lista_randomica)
    lista_media.append(media)
    i = i + 1
```

```python
# Criação do Histograma
plt.hist(lista_media)
plt.show()
```
