# Exercício começando a entender estatística

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Faz o carregamento do arquivo csv
dataset = open('./pokemon.csv', encoding='utf-8')
dataset_csv = csv.reader(dataset, delimiter= ',', quotechar = '"')

header = next(dataset_csv)
header_position = {}

for i in range(len(header)):
    header_position[header[i]] = i

weight_posi = header_position['weight_kg']
pokemon_weight_list = []

for line in dataset_csv:
    weight = line[weight_posi]
    if weight != '':
        weight = float(weight)
        pokemon_weight_list.append(weight)
print(pokemon_weight_list)
```

```python
i = 0
list_randomica = {}
list_media = []

# Gerando um for para as 1000 vezes que a média será gerada
for i in range(1000):
    # Gerando amostras aleatórias com o número de 20 elementos 
    list_randomica = random.sample(pokemon_weight_list, 20)
    # Gerando a média
    media = np.mean(list_randomica)
    # Adicionando as médias dos dados randomicos para a geração do histograma
    list_media.append(media)
    i = i + 1
```

```python
# Gerando um histograma das médias
plt.hist(list_media)
plt.show()
```
